# Machine Learning in Python
- Supervised learning is a branch of machine learning.
- It uses labelled data to learn and later on predicts unlabelled data based on the acquired knowledge.
- Python through scikit-learn or 'sklearn' provides a wide range of suprevised learning algorithm for classification and regression problems.

```python
from sklearn import datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = datasets.load_breast_cancer()
data.keys()
```

```python
# The data/observation entries are stored as values in the 'data' key.
X = pd.DataFrame(data['data'])

# The target entries/labels are stored as values in 'target' key.
y = pd.DataFrame(data['target'])
```

```python
# Set the valeues of 'feature_names' as the column names
X.columns = data['feature_names']
```

### Perform kNN Algorithm


```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_recall_fscore_support
from scikitplot.metrics import plot_confusion_matrix as plt_confusion
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=31)
```

```python
len(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier

#Instantiate the KNN classfier
knn = KNeighborsClassifier(n_neighbors=3)

#Train the model
#Flatten y_train array with ravel() to avoid warnings - optional
knn.fit(X_train, y_train.values.ravel())
```

```python
#Predict using the trained model
prediction = knn.predict(X_test)
prediction_prob = knn.predict_proba(X_test) #calculate prediction probability
```

```python
prediction_prob
```

```python
# Evaluate the accuracy of the model
knn_score = knn.score(X_test, y_test.values.ravel())
```

```python
knn_score # Served as the baseline accuracy
# If want to improve the model, at least more than 90.64%
```

- The accuracy of kNN is 90.64%

```python
# Plot confusion matrix using the plt_confusion method imported earlier
plt_confusion(y_test, prediction)
```

```python
# Calculate the evaluation metrics using precision_recall_fscore_support method
scoring = precision_recall_fscore_support(y_test, prediction) # Calculate alternative metrics

scoring = pd.DataFrame(list(scoring))
scoring.columns = data.target_names
scoring.index = ['Precision', 'Recall', 'F-Score', 'Support']
np.round(scoring, decimals=2) # round each entry in data frame to 2 decimal places

print('Evaluation metrics for kNN is:')
scoring
```

```python
# Plot the ROC curve using the plt_roc method imported earlier
import scikitplot as scp
scp.metrics.plot_roc(y_test, prediction_prob, title='kNN ROC Curve', figsize=(8,8))
plt.show()
```

### K-Fold Cross Validation
- We can try enhancing the model's performance using cross validation.
- Cross validation increases the amount of training data that your model fits and may substantially increase its performance.
- Perform K-Fold Cross Validation using cross_val_score from sklearn.model_selection module.
- Set the number of folds to 10 via the cv parameter.

```python
from sklearn.model_selection import cross_val_score

# Flatten y array using ravel() to avoid warning - optional
cv_results_knn = cross_val_score(knn, X, y.values.ravel(), cv = 10)
cv_results_rf = cross_val_score(rf, X, y.values.ravel(), cv = 10)

print('The 10-fold cross valudation result for kNN is {:.2f}%'.format(np.mean(cv_results_knn)*100))
print('The 10-fold cross valudation result for Random Forest is {:.2f}%'.format(np.mean(cv_results_rf)*100))
```

***

### Random Sampling Cross Validation
- Another cross validation approach is by using random sub-sampling.
- This algorithm randomly selectes a portion of our data to be the training set and the remaining portion as the test set.
- This process repeats for a specific amount of iterations as configured.
- Use the ShuffleSplit method from sklearn.model_selection module.
- Set the number of iterations to 10 via the n_splits parameter.
- Set test_size = 0.3 to configure the test size portion to be 0.3 (30%)

```python
# Perform Random Sampling Cross Validation
from sklearn.model_selection import ShuffleSplit

# Instantiate Random Sampling CV Object
rs = ShuffleSplit(n_splits=10, test_size=0.3, random_state=10)

# Create empty lists to store score for kNN and Random Forest
randomcv_knn = []
randomcv_rf = []

# Generate train and test index using random splitting via the split() method
# Flatten array using ravel() to avoid warnings - optional
for train_index, test_index in rs.split(X):
    knn.fit(X.loc[train_index], y.loc[train_index].values.ravel())
    rf.fit(X.loc[train_index], y.loc[train_index].values.ravel())
    
    # Append scoring/accuracy for each iteration in list
    randomcv_knn.append(knn.score(X.loc[test_index], y.loc[test_index].values.ravel()))
    randomcv_rf.append(rf.score(X.loc[test_index], y.loc[test_index].values.ravel()))
    
# Calculate the average of all iteration
print('The random sampling cross validation result ({} iterations) for kNN is {:.2f}%'
      .format(rs.get_n_splits(X), np.mean(randomcv_knn)*100))
print('The random sampling cross validation result ({} iterations) for Random Forest is {:.2f}%'
      .format(rs.get_n_splits(X), np.mean(randomcv_rf)*100))
```

### Leave-One-Out Cross Validation
- The algorithm takes the entire dataset except for one to be the training set.
- The remaining sample left out will be the testing set.
- This algorithm is really expensive if you have large amount of data.
- This is because it runs on every sample of your entry. 
- For example, if you have 10 data, the algorithm takes 9 to be the training set and 1 to be the testing set.
- It then progresses to next sample until the end.
- Use the LeaveOneOut method from sklearn.model_selection module.

```python
# Perform Leave-One-Out Cross Validation
from sklearn.model_selection import LeaveOneOut

# Instantiate LeaveOneOut CV object
loo = LeaveOneOut()

# Create empty lists to store score for kNN and RF
loocv_knn = []
loocv_rf = []

# Generate train and test index using random splitting via the split() method
# Flatten array using ravel() to avoid warnings - optional
for train_index, test_index in loo.split(X):
    knn.fit(X.loc[train_index], y.loc[train_index].values.ravel())
    rf.fit(X.loc[train_index], y.loc[train_index].values.ravel())
    
    # Append scoring/accuracy for each iteration in list
    loocv_knn.append(knn.score(X.loc[test_index], y.loc[test_index].values.ravel()))
    loocv_rf.append(rf.score(X.loc[test_index], y.loc[test_index].values.ravel()))
    
# Calculate the average of all iteration
print('The leave-one-out cross validation result for kNN is {:.2f}%'.format(np.mean(loocv_knn)*100))
print('The leave-one-out cross validation result for Random Forest is {:.2f}%'.format(np.mean(loocv_rf)*100))
```

```python
accuracies = pd.DataFrame(data=
                    {'Cross Validation Methods': ['K-Fold', 'Random Sampling', 'Leave-One-Out'], 
                     'k-Nearest Neighbours': (np.mean(cv_results_knn)*100, np.mean(randomcv_knn)*100, np.mean(loocv_knn)*100), 
                     'Random Forest': [np.mean(cv_results_rf)*100, np.mean(randomcv_rf)*100, np.mean(loocv_rf)*100]})

accuracies
```

- We can see that Random Forest algorithm has a better accuracy on the data compared to kNN.
- Also, among the cross-validation methods, random sampling cross validation has the most accurate score.
